```python
import pandas as pd
```

```python
SAMPLE_SIZE = 500
OUTPUT_FILE = 'sample_master.csv'
```

```python
df_model = pd.read_csv('blind_pool_cross_model.csv')
df_model.head(1)
```

```python
df_prompt = pd.read_csv('blind_pool_cross_prompt.csv')
df_prompt.head(1)
```

```python
df_lingual = pd.read_csv('blind_pool_cross_lingual.csv')
df_lingual.head(1)
```

```python
input_file = [df_prompt, df_model, df_lingual]
master_samples = []

for df in input_file:

    df_sample = df.sample(n=SAMPLE_SIZE, random_state=42).copy()
    df_sample = df_sample.reset_index(drop=True)

    master_samples.append(df_sample)

len(master_samples)
```

```python
df_master = pd.concat(master_samples, ignore_index=True)
df_master.shape
```

```python
df_master.head()
```

```python
ESSENTIAL_COLS = [
    'evaluation_id', 'response_A', 'response_B'
]
```

```python
df_essentials = df_master[ESSENTIAL_COLS].copy()
df_essentials.head()
```

```python
df_master.to_csv(OUTPUT_FILE, index=False, encoding='utf-8-sig')
print(f"Master sample created in '{OUTPUT_FILE}' with {len(df_master)} pairs.")
```
